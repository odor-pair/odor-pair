```python
import json
trial_dir = "paperdata/trials"
trial_fnames = os.listdir(trial_dir)
trials = []
for trial_fname in trial_fnames:
    if not trial_fname.endswith(".json"):
        continue
    with open(os.path.join(trial_dir,trial_fname)) as f:
        trials.append(json.load(f))
len(trials),trials[0]
```

```python
# scores = [trial["final"] for trial in trials]
scores = [max(trial["results"].values()) for trial in trials]
scores
```

```python
import numpy as np
import scipy.stats as stats

# Step 1: Calculate the mean
mean = np.mean(scores)

# Step 2: Calculate the standard error of the mean (SEM)
sem = stats.sem(scores)

# Step 3: Get the t-critical value for 95% confidence with df = len(scores) - 1
confidence = 0.95
t_critical = stats.t.ppf((1 + confidence) / 2, df=len(scores) - 1)

# Step 4: Calculate the margin of error
margin_of_error = t_critical * sem

# Step 5: Calculate the confidence interval
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

# Print the result
print(f"95% Confidence Interval: {confidence_interval}")
print(f"Mu {mean} +- {margin_of_error}")
```
