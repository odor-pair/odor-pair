```python
# Find the set of keys common to all dictionaries
common_keys = set(scores[0].keys()).intersection(*[d.keys() for d in scores])

# Display the common keys
print(common_keys)
```

```python
# Convert the list of dictionaries to a DataFrame using only the common keys
df = pd.DataFrame([{k: d[k] for k in common_keys} for d in scores], index=["MPNN-GNN", "GIN-GNN","Morgan Fingerprint"])

# Calculate the average rank order for each key
# sorted_keys = df.mean(axis=0).map(lambda x: 1/x).sort_values()
sorted_keys = df.rank(axis=1, ascending=False, method='average').mean().sort_values()
```

```python
best_notes = sorted_keys.nsmallest(10)
print(best_notes)
print()
worst_notes = sorted_keys.nlargest(10).sort_values(ascending=False)
print(worst_notes)
```

```python
top_10_df = df[best_notes.index].map(lambda x: f'{x:.2f}').T
top_10_df
```

```python
import matplotlib.pyplot as plt

def make_table(top_df,title,fname):
    # Create a matplotlib table for the top 10 keys
    fig, ax = plt.subplots(figsize=(5.5,2.5))
    table = ax.table(cellText=top_df.values,
                      colLabels=top_df.columns,
                      rowLabels=top_df.index,
                      loc='center',
                      cellLoc='center')
    table.auto_set_font_size(False)
    ax.axis('off')  # Hide axes
    plt.title(title)
    plt.tight_layout()
    # plt.savefig(fname)
    plt.show()
make_table(top_10_df,"Top 10 Predicted Labels, by Model","Top10.png")
```

```python
bottom_10_df = df[worst_notes.index].map(lambda x: f'{x:.2f}').T
bottom_10_df
```

```python
make_table(bottom_10_df,"Bottom 10 Predicted Labels, by Model","Bottom10.png")
```

```python
import matplotlib.pyplot as plt

def make_chart(top_df,title,fname):
    # Create a matplotlib table for the top 10 keys
    fig, ax = plt.subplots()
    top_df = top_df.map(lambda x: float(x))
    top_df.plot(kind='bar', ax=ax)
    plt.title(title)
    ax.tick_params(axis='x', rotation=45)
    plt.tight_layout()
    utils.save_figures(fname,fig)
    plt.show()
make_chart(top_10_df,"Top 10 Predicted Labels, by Model","Top10Bar")
```
