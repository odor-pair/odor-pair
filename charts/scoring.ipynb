```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size': 12})

def make_multi_row_performance(task,fname,aurocs,df,rows,bbox_to_anchor,hspace=.5, ):
    df = df.copy(deep=True).drop(['TOTAL'], axis=1, errors='ignore')
    counter_per_row = int(np.ceil(len(df.columns)/rows))
    fig, axs = utils.standard_fig_ax(x=rows)
    fig.subplots_adjust(hspace=hspace)  # Increase vertical spacing
    w = (1 / len(df.index)) - .08

    column_chunks = [df.columns[i:i + counter_per_row] for i in range(0, len(df.columns), counter_per_row)]
    for i, columns in enumerate(column_chunks):
        for j,m in enumerate(df.index):
            x_offsets = np.arange(len(columns)) + w * j  # Offsetting each bar by 0.1 per row
            axs[i].bar(x_offsets, df.loc[m, columns], width=w, align="edge",label=m)

        axs[i].axhline(y=0.5,color='grey',linestyle='dashed')
        axs[i].tick_params(axis='x', rotation=35)
        axs[i].set_ylim(0,1)
        axs[i].set_xticks(ticks=np.arange(len(columns))+w*len(df.index)/2,labels=columns)
        for label in axs[i].get_xticklabels():
            label.set_ha('right')  # 'ha' is shorthand for 'horizontalalignment'

        
    # Get the handles from any axis
    handles, _ = axs[-1].get_legend_handles_labels()
    labels = [f"{m} (AUROC={aurocs[m]:.3f})" for m in df.index]
    legend = plt.figlegend(handles, labels, bbox_to_anchor=bbox_to_anchor)

    plt.suptitle(f"AUROC Comparison on {task} Task\nby Model and Odor Label",y=.93)
    fig.set_size_inches(15,10)
    utils.save_figures(fname,fig)
    plt.show()
    
aurocs = {m:calc_micro_average(m) for m in sorted_df.index}
make_multi_row_performance("Blend Pair","pairall_3rows",aurocs, sorted_df,3,bbox_to_anchor=(.72,.87,.2,.1))
```

```python
import os
import json

# Define the path to the results folder
results_folder = 'mpnn_gnn_results'

# Initialize variables to keep track of the file with the highest TOTAL
max_total = None
max_total_file = None

# Iterate over all files in the results folder
for file_name in os.listdir(results_folder):
    # Check if the file is a JSON file
    if file_name.endswith('.json'):
        # Construct the full file path
        file_path = os.path.join(results_folder, file_name)
        if not "single" in file_path:
            continue
        # Open and read the JSON file
        with open(file_path, 'r') as json_file:
            data = json.load(json_file)
            # Print the file name and the TOTAL value
            print(f"File: {file_name}, TOTAL: {data.get('TOTAL')}")
           # Get the TOTAL value
            total = data.get('TOTAL')
            # Check if this is the highest TOTAL we've seen so far
            if max_total is None or (total is not None and total > max_total):
                max_total = total
                max_total_file = file_name
print(f"File with the highest TOTAL: {max_total_file}, TOTAL: {max_total}")
```
